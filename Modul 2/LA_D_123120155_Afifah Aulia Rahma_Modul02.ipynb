<p>1.3 Teori Dasar</p>
    <p style="text-align: justify;">
    Gaya Berat atau metode gaya berat merupakan salah satu metode dalam eksplorasi geofisika yang digunakan untuk mengukur variasi medan gravitasi bumi akibat adanya perbedaan densitas antar batuan. Proses pengukuran gaya berat memiliki tiga tahapan, yaitu - pengukuran - pengolahan data, dan - interpretasi hasil dari pengolahan data yang telah dilakukan. Ada beberapa koreksi yang harus dilakukan saat pengumpulan data gaya berat, sehingga kita akan mendapatkan nilai yang sebenarnya. Koreksi yang dipakai pada metode gaya berat salah satunya adalah Koreksi Tidal (pasang surut bumi) dan Koreksi Drift (apungan). 
</p>
    <p style="text-align: justify;">
        Koreksi Medan (Terrain Correction) merupakan koreksi yang dilakukan akibat
pendekatan Bouguer. Topografi bumi tidaklah datar seperti slab yang digunakan pada
koreksi bouguer (BC) tetapi berundulasi. Topografi sekitar mempengaruhi nilai gravitasi
pada saat pengukuran baik bukit maupun lembah. Perhitungan TC dapat menggunakan
Hammer chart. Berdasarkan radiusnya koreksi ini dibagi menjadi 2, yaitu inner zone dan
outer zone.
<li>1. Inner Zone</li>
<p> Memiliki radius yang tidak terlalu besar sehingga bias didapatkan dari pengamatan
langsung di lapangan. Terdiri dari zona B (radius 6.56 ft dibagi 4 sektor) dan zona C
(radius 54.6 ft dibagi 6 sektor). </p>
<li>2. Outer Zone</li>
<p>Zona ini memiliki radius yang cukup sehingga bisanya perbedaan ketinggian dengan
titik pengukuran gravity menggunakan analisis peta kontur. Outer zone dibagi menjadi
beberapa zona yaitu zona D (radius 175 ft dibagi 6 sektor), zona E (radius 558 ft dibagi
8 sektor), zona F (radius 1280 ft dibagi 8 sektor), zona G (radius 2936 ft dibagi 12
sektor), zona H (radius 5018 ft dibagi 12 sektor), zona I (radius 8575 ft dibagi 12 sektor), zona J (radius 14612 ft dibagi 12 sektor), zona K sampai M (masing-masing dibagi 6
sektor).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from pyproj import Proj, transform
```

Masukkan data awal

```python
df = pd.read_csv("D:\smt 5\Metode Gayaberat\minggu 1\LA\modul 1\Data input\Data input\Data_Awalyangdimasukan.txt",delim_whitespace=True)

print(df.to_string(index=False))
```

Menampilkan Data Waktu dan Tanggal untuk koreksi tidal

```python
df_waktu_tanggal = df[['Waktu', 'Tanggal']]
print(df[['Waktu', 'Tanggal']])

output_txt = "Waktu_Tanggal.txt"
df_waktu_tanggal.to_csv(output_txt, sep="\t", index=False)

print (f" Data Waktu dan Tanggal berhasil disimpan ke file: {output_txt}")
```

Mengubah waktu ke menit

```python
def waktu_ke_menit(waktu_str):
  try:
    jam, menit, _ = waktu_str.strip().split('.')
    return int (jam) * 60 + int (menit)
  except Exception as e:
      print(f"error saat mengkonversi waktu: {e}")
      return None
df['Menit'] = df['Waktu'].apply(waktu_ke_menit)
print(df[['Tanggal', 'Waktu', 'Menit']].to_string(index=False))
```

koreksi tidal

```python
Tidal = pd.read_csv('D:\smt 5\Metode Gayaberat\minggu 1\LA\modul 1\Data input\Data input\Tidal_Correction.txt', delimiter='\t')
if 'Tidal' not in Tidal.columns:
    raise ValueError("Kolom 'Tidal' tidak ditemukan dalam file Tidal_Correction.txt!")

if len(df) >= len(Tidal):
    df['tide'] = Tidal['Tidal'].values.tolist() + [None] * (len(df) - len(Tidal))
    df['Gtide'] = df['Gobs(mGal)'] + df['tide']
else:
    raise ValueError("Jumlah baris Tidal_Correction.txt lebih banyak dari Data_Awal.txt!")

print(df[['X', 'Y', 'Z', 'Waktu', 'Tanggal', 'Gobs(mGal)', 'tide', 'Gtide']].to_string(index=False))
```

Koreksi Drift

```python
df['Tanggal'] = pd.to_datetime(df['Tanggal'], format='%d-%m-%Y')

def hitung_drift(row, df):
    tanggal = row['Tanggal']
    menit_stasiun = row['Menit']

    df_tanggal = df[df['Tanggal'] == tanggal]

    if not df_tanggal.empty:
        Gtide_awal = df_tanggal.iloc[0]['Gtide']
        Gtide_akhir = df_tanggal.iloc[-1]['Gtide']
        menit_awal = df_tanggal.iloc[0]['Menit']
        menit_akhir = df_tanggal.iloc[-1]['Menit']

        drift = (Gtide_awal - Gtide_akhir) * (menit_stasiun - menit_awal) / (menit_akhir - menit_awal)
        return drift
    return None

df['Drift'] = df.apply(lambda row: hitung_drift(row, df), axis=1)
df['Gdrift'] = df['Gtide'] + df['Drift']

print(df[['Tanggal', 'Waktu', 'Menit', 'Gtide', 'Drift', 'Gdrift']].to_string(index=False))
```

Input Nilai Gbase

```python
Gbase = 978129.97
print ('Gbase:', Gbase)
```

Menghitung Gobs Hasil Koreksi

```python
df['Gdrift_awal'] = df.groupby('Tanggal')['Gdrift'].transform('first')
df['Delta_G'] = df['Gdrift'] - df['Gdrift_awal']
df['Gobs_HasilKoreksi'] = Gbase + df['Delta_G']

print("\n=== Data dengan Delta G dan Gobs Hasil Koreksi ===\n")
print(df[['Tanggal', 'Z', 'Waktu', 'Menit', 'Gdrift', 'Gdrift_awal', 'Delta_G', 'Gobs_HasilKoreksi']].to_string(index=False))
```

Mengitung Perbandingan Gobs dengan Htrue

```python
df['No_Titik'] = range(1, len(df)+1)

#Kurva Z
plt.figure(figsize=(30,5))
plt.plot(df['No_Titik'], df['Z'], marker='o', linestyle='-', color='green')
plt.title("Kurva Elevasi (Z) terhadap Titik Pengukuran")
plt.xlabel("Nomor Titik Pengukuran")
plt.ylabel("Elevasi (Z)")
plt.grid(True)
plt.show()

#Kurva Gobs hasil koreksi
plt.figure(figsize=(30,5))
plt.plot(df['No_Titik'], df['Gobs_HasilKoreksi'], marker='s', linestyle='-', color='blue')
plt.title("Kurva Gobs Hasil Koreksi terhadap Titik Pengukuran")
plt.xlabel("Nomor Titik Pengukuran")
plt.ylabel("Gobs Hasil Koreksi (mGal)")
plt.grid(True)
plt.show()
```

Mengubah Latitude (RAD)

```python
def hitung_latitude_rad(y):
  return np.radians(y)

df['Latitude_rad'] = df['Y'].apply(hitung_latitude_rad)

print(df[['Y', 'Latitude_rad']].to_string(index=False))
```

Koreksi Lintang

```python
def hitung_glintang(latitude_rad):
    return 978032.7 * (1 + 0.0053024 * np.sin(latitude_rad)**2 - (0.0000058 * np.sin(2 * latitude_rad)**2))

df['Glintang'] = df['Latitude_rad'].apply(hitung_glintang)
print(df[['Latitude_rad', 'Glintang']].to_string(index=False))
```

FAC

```python
def hitung_fac(z):
  return 0.3085 * z

df['FAC'] = df['Z'].apply(hitung_fac)
print(df[['Z', 'FAC']].to_string(index=False))
```

Koreksi FAA (Free Air Anomaly)

```python
def hitung_faa(gobs_koreksi, glintang, fac):
    return gobs_koreksi - glintang + fac

df['FAA'] = df.apply(lambda row: hitung_faa(row['Gobs_HasilKoreksi'], row['Glintang'], row['FAC']), axis=1)
print(df[['Gobs_HasilKoreksi', 'Glintang', 'FAC', 'FAA']].to_string(index=False))
```

Menampilkan Hasil lintang FAC dan FAA

```python
print(df[['Tanggal', 'Waktu', 'X', 'Y', 'Z', 'Menit', 'Latitude_rad','Glintang','FAC','FAA']].to_string(index=False))
```

Eliminasi Nilai awal dan Akhir pada masing masing Tanggal

```python
df['is_first'] = df.groupby('Tanggal')['Gobs_HasilKoreksi'].transform('first') == df['Gobs_HasilKoreksi']
df['is_last']  = df.groupby('Tanggal')['Gobs_HasilKoreksi'].transform('last')  == df['Gobs_HasilKoreksi']

df_filtered = df[~(df['is_first'] | df['is_last'])].copy()

df_filtered = df_filtered.drop(columns=['is_first','is_last'])

print(df_filtered[['Tanggal','Waktu','Menit','Gobs_HasilKoreksi']].to_string(index=False))
```

Mencari nilai BC (asumsi densitas rata-rata bumi = (2.67))

```python
# =2,67

df_filtered.dropna(subset=['Z'], inplace=True)
df_filtered['BC'] = 2 * np.pi * 6.674*10**-11*2.67*1000 * df_filtered['Z'] * 10**5

print(df_filtered[['Tanggal','Waktu','Menit','Z','BC']].to_string(index=False))
```

SBA

```python
df_sba = df_filtered.copy()

df_sba['SBA']=df_sba['Gobs_HasilKoreksi']-df_sba['Glintang'] + df_sba['FAC'] - df_sba['BC']
print(df_sba[['Glintang', 'FAC', 'BC','SBA']].to_string(index=False))
```

Menampilkan long, lat, dan SBA

```python
df_output = df_sba.copy()
hasil_SBA = pd.DataFrame({
    'X': df_filtered['X'].iloc[:len(df_output)].values,
    'Y': df_filtered['Y'].iloc[:len(df_output)].values,
    'SBA': df_output['SBA'].values
})

print(hasil_SBA.to_string(index=False))
```

Konversi ke utmx dan utmy

```python
proj_wgs84 = Proj(proj="latlong", datum="WGS84")
proj_utm = Proj(proj="utm", zone=48, datum="WGS84", south=True)

longitudes = hasil_SBA['X'].values
latitudes = hasil_SBA['Y'].values

utmx, utmy = transform(proj_wgs84, proj_utm, longitudes, latitudes)

hasil_SBA_UTM = pd.DataFrame({
    'UTMX': utmx,
    'UTMY': utmy,
    'SBA': hasil_SBA['SBA'].values
})

print(hasil_SBA_UTM.to_string(index=False))
```

Menyimpan koordinat UTmx, UTmy dan Nilai SBA

```python
hasil_SBA_UTM.to_csv("filehasilpengolahanSBA.txt", sep="\t", index=False)

print("\n Hasil berhasil disimpan ke 'filehasilpengolahanSBA.txt'")
```

Menampilkan peta SBA dan menyimpannya dalam tif

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import Rbf
import rasterio
from rasterio.transform import from_origin
from matplotlib import cm

data = pd.read_csv("filehasilpengolahanSBA.txt", sep="\t")
x = data['UTMX'].values
y = data['UTMY'].values
z = data['SBA'].values

grid_res = 500
xi = np.linspace(x.min(), x.max(), grid_res)
yi = np.linspace(y.min(), y.max(), grid_res)
xi, yi = np.meshgrid(xi, yi)


rbf = Rbf(x, y, z, function='thin_plate')
zi = rbf(xi, yi)

plt.figure(figsize=(15, 10))
contour = plt.contourf(xi, yi, zi, cmap='seismic', levels=200)
plt.scatter(x, y, c='black', s=80, edgecolors='white', linewidths=0.5, label='Titik Data')
plt.colorbar(contour, label='CBA (mGal)')
plt.xlabel('UTM X (m)')
plt.ylabel('UTM Y (m)')
plt.title('Peta Kontur SBA - Interpolasi RBF')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

norm = plt.Normalize(vmin=np.nanmin(zi), vmax=np.nanmax(zi))
colormap = cm.get_cmap('seismic')
rgba_img = colormap(norm(zi))
rgb_img = (rgba_img[:, :, :3] * 255).astype(np.uint8)

rgb_img_flipped = np.flipud(rgb_img)


pixel_size_x = (x.max() - x.min()) / (rgb_img.shape[1] - 1)
pixel_size_y = (y.max() - y.min()) / (rgb_img.shape[0] - 1)
transform = from_origin(x.min(), y.max(), pixel_size_x, pixel_size_y)

with rasterio.open(
    'peta_SBA_kelompok1_RBF_RGB.tif',
    'w',
    driver='GTiff',
    height=rgb_img_flipped.shape[0],
    width=rgb_img_flipped.shape[1],
    count=3,
    dtype='uint8',
    crs='EPSG:32748',
    transform=transform,
) as dst:
    for i in range(3):
        dst.write(rgb_img_flipped[:, :, i], i + 1)
```

<div style="text-align: center; font-family: 'Times New Roman', Times, serif;">
    <h1>Analisis</h1>
</div>

<div style="font-family: 'Times New Roman', Times, serif; margin: 20px auto; width: 80%;">
    <p style="text-align: justify;">
Dalam melakukan pengolahan data gaya berat, data yang dihasilkan merupakan data yang sudah dikoreksi atau data yang sudah dihilangkan dari pengaruh atau factor disekitarnya. Secara umum, metode ini dapat dimanfaatkan dalam berbagai bidang seperti eksplorasi panas bumi, eksplorasi mineral, minyak dan gas, identifikasi patahan dan struktur patahan, penelitian air tanah, cekungan hidrokarbon, dan lain-lain. Metode gaya berat adalah salah satu metode dalam geofisika yang memiliki tingkat ambiguitas yang cukup tinggi. Ketika pengukuran menggunakan alat, data yang diperoleh masih dipengaruhi oleh beberapa faktor tertentu. Padahal, data yang kita butuhkan adalah nilai yang telah bebas dari pengaruh faktor-faktor tersebut. Oleh karena itu, diperlukan koreksi data gaya berat untuk menghilangkan pengaruh faktor tersebut. Secara umum pengukuran dengan metode gravity dilakukan untuk mengetahui nilai densitas pada daerah penelitian tersebut. 
</p>

<p>
Secara berurutan, tahapan dalam pengolahan di colab diawali dengan import modul pengolahan data seperti numpy, pandas, dan matplotlib, kemudian pembacaan data gravitasi lapangan dalam format .txt. Data tersebut berisi kolom waktu, tanggal, lintang, bujur, dan elevasi. Tahap awal meliputi konversi waktu ke satuan menit, sesuai langkah ke-5 modul, menggunakan fungsi waktu_ke_menit(). Tahapan berikutnya adalah perhitungan koreksi pasang surut (tide correction) untuk menghilangkan pengaruh bulan dan matahari, serta koreksi drift untuk menetralkan penyimpangan alat akibat waktu pengukuran yang berbeda. Koreksi drift dihitung menggunakan fungsi hitung_drift().</p>
<p>
Setelah seluruh koreksi awal dilakukan, nilai Gobs hasil koreksi diperoleh melalui kombinasi nilai Gbase, Gdrift_awal, dan G. Hasilnya kemudian digunakan untuk menghitung anomali udara bebas (FAA) dengan rumus FAA = Gobs_HasilKoreksi  Glintang + FAC, di mana Glintang dihitung berdasarkan rumus GRS80 dan FAC dihitung menggunakan gradien 0.308765 mGal/m. Langkah selanjutnya adalah koreksi Bouguer (BC) dengan rumus BC = 0.04188    h ( = 2.67 g/cm), sehingga diperoleh nilai SBA. Data SBA kemudian dikonversi ke sistem koordinat UTM (zona 48S) untuk pembuatan peta kontur. 
</p></div>

<div style="font-family: 'Times New Roman', Times, serif; margin: 20px auto; width: 80%;">
    <p style="text-align: justify;">
Peta SBA dengan metode Minimum Curvature menggambarkan distribusi nilai Simple Bouguer Anomaly hasil interpolasi dari data gravitasi yang telah dikoreksi. Pola warna pada peta menunjukkan variasi densitas bawah permukaan. Warna biru hingga ungu merepresentasikan nilai anomali negatif yang mengindikasikan keberadaan zona dengan massa jenis rendah, seperti batuan sedimen, rekahan, atau daerah yang memiliki banyak struktur geologi. Sebaliknya, warna merah hingga magenta menunjukkan nilai anomali positif yang menandakan adanya batuan dengan densitas tinggi, seperti batuan beku atau intrusi di bawah permukaan. Metode Minimum Curvature digunakan dengan tujuan menghasilkan permukaan interpolasi yang halus dan kontinu antar titik data, mengikuti prinsip energi minimum sehingga kontur tidak kontras antar area. Secara pengolahan, peta ini merupakan hasil tahap akhir dari seluruh koreksi data gaya berat sesuai langkah-langkah pada Modul 1 Perhitungan SBA. Setelah dilakukan koreksi pasang surut, koreksi drift, koreksi lintang, koreksi udara bebas, serta koreksi Bouguer, nilai SBA dihitung menggunakan rumus SBA = Gobs_koreksi  Glintang + FAC  BC. Nilai SBA yang diperoleh kemudian diinterpolasi dalam sistem koordinat UTM menggunakan metode Minimum Curvature, hal ini dilakukan untuk membentuk grid kontur yang sama. Hasil interpolasi divisualisasikan dalam bentuk peta kontur berwarna. Peta SBA Minimum Curvature  berfungsi untuk menggambarkan kondisi regional anomali gravitasi di area penelitian. Karena metode ini menghasilkan kontur yang halus, hasilnya bagus untuk melihat pola besar seperti punggungan anomali tinggi dan lembah anomali rendah. Hal ini sesuai dengan teori dasar  yang menjelaskan bahwa anomali Bouguer sederhana menunjukkan variasi densitas batuan di bawah permukaan akibat perbedaan sifat fisik geologi, sehingga interpretasi dapat diarahkan pada identifikasi struktur bawah permukaan. </p></div>

<div style="font-family: 'Times New Roman', Times, serif; margin: 20px auto; width: 80%;">
    <p style="text-align: justify;">
Peta SBA dengan metode Kriging menampilkan distribusi anomali Bouguer sederhan. Berbeda dengan Minimum Curvature, pada metode Kriging hubungan spasial antar titik data juga dipertimbangkan melalui variogram, sehingga interpolasi yang dihasilkan lebih terkontrol dan sesuai dengan pola penyebaran data pengukuran. Pola kontur pada peta menunjukkan transisi warna yang teratur, di mana warna biru hingga hijau menandakan anomali negatif, sedangkan warna kuning hingga magenta menandakan anomali positif. Dari sisi pengolahan, peta ini juga berasal dari data SBA hasil koreksi lengkap sesuai tahapan modul praktikum, namun interpolasinya menggunakan algoritma Kriging. Prosesnya mulai dari pembacaan nilai UTMX, UTMY, dan SBA, kemudian dilakukan grid data dengan resolusi tertentu. Interpolasi Kriging menghitung nilai setiap titik grid berdasarkan kombinasi bobot dari titik pengamatan di sekitarnya, sehingga menghasilkan model yang realistis secara spasial. Peta ini menggambarkan distribusi densitas bawah permukaan dengan tingkat detail lebih tinggi dibandingkan metode Minimum Curvature, karena mempertimbangkan variasi lokal yang lebih sensitif. Pada interpretasinya, peta SBA Kriging membantu mengidentifikasi batas-batas zona anomali. Area berwarna biru menunjukkan zona densitas rendah yang berpotensi sebagai endapan sedimen, sedangkan zona berwarna merah hingga magenta menunjukkan densitas tinggi, yang kemungkinan berasosiasi dengan batuan intrusi atau basement. 
</p></div>

<div style="font-family: 'Times New Roman', Times, serif; margin: 20px auto; width: 80%;">
    <p style="text-align: justify;">
Peta overlay SBA Minimum Curvature merupakan hasil gabungan antara peta anomali Bouguer sederhana hasil interpolasi Minimum Curvature dengan peta geologi regional. Tujuan overlay ini adalah untuk menghubungkan nilai anomali gravitasi dengan satuan geologi permukaan guna memperoleh interpretasi geologi yang mendetail. Berdasarkan peta overlay, zona anomali tinggi (merah hingga magenta) berada di wilayah dengan satuan batuan Formasi Tomh dan Pag yang didominasi oleh batuan beku dan batuan keras. Hal ini menunjukkan bahwa nilai gravitasi meningkat akibat densitas batuan yang tinggi. Sebaliknya, zona anomali rendah (biru) berada di area dengan satuan geologi QTI dan Tmgr yang merupakan endapan muda dan sedimen lepas, sehingga densitasnya lebih rendah. Pembuatan peta overlay ini dilakukan setelah hasil grid SBA Minimum Curvature disimpan dalam format GeoTIFF. Data itu kemudian diimpor ke QGIS, lalu dioverlay dengan peta geologi menggunakan sistem koordinat yang sama yaitu WGS 84/UTM zona 48S. Hasil overlay menampilkan hubungan spasial antara anomali gravitasi dan litologi permukaan, yang mendukung interpretasi bahwa variasi nilai gravitasi dipengaruhi oleh jenis batuan penyusunnya.Variasi anomali Bouguer sederhana mencerminkan variasi densitas batuan di bawah permukaan, zona dengan anomali positif cenderung dikaitkan dengan batuan beku atau intrusi, sedangkan zona negatif umumnya disebabkan oleh batuan sedimen yang lebih ringan atau daerah yang mengalami pelapukan. Metode Minimum Curvature bekerja dengan prinsip meminimalkan energi kelengkungan total pada permukaan grid hasil interpolasi. Oleh karena itu, peta yang dihasilkan menampilkan kontur yang sangat halus dan kontinyu, dengan transisi warna yang natural dan tidak menimbulkan perubahan mendadak antar area. 
</p></div>

<div style="font-family: 'Times New Roman', Times, serif; margin: 20px auto; width: 80%;">
    <p style="text-align: justify;">
Peta overlay SBA Kriging menampilkan hasil interpolasi Kriging yang dioverlay dengan peta geologi pada titik pengukuran yang sama. Pola distribusi anomali pada peta ini mirip dengan hasil Minimum Curvature, nadengan peerbedaan batas-batas zona anomali yang terlihat lebih tajam dan sejajar dengan batas satuan geologi. Kelebihan metode Kriging terletak pada kemampuan untuk mempertahankan nilai data di sekitar titik pengamatan, sehingga hasil overlay memberikan gambaran yang lebih akurat mengenai variasi densitas bawah permukaan terhadap litologi yang ada. Pada peta ini, zona anomali tinggi masih berasosiasi dengan batuan keras dan tua seperti Formasi Pag dan Tomg, sedangkan zona anomali rendah tetap muncul pada area dengan batuan sedimen muda seperti Qa dan QTI. Hasil ini menunjukkan konsistensi pola antara peta SBA Kriging dan kondisi geologi setempat. Penggunaan Kriging memungkinkan interpretasi struktur bawah permukaan yang lebih presisi, seperti penentuan posisi sesar atau batas formasi batuan. Interpolasi kriging memperhitungkan hubungan spasial antar titik data. Secara pengolahan, langkah pembuatan overlay hampir sama dengan metode Minimum Curvature, dimulai dari ekspor hasil grid SBA ke format GeoTIFF, kemudian dikombinasikan dengan peta geologi di sistem koordinat WGS84/UTM zona 48S. Hingga akhirnya menampilkan kesamaan antara anomali gravitasi dan karakteristik geologi.
    </p>
</div>
